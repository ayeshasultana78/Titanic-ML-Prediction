```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
# Load Titanic
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)
print(df.shape)
print(df.head())
```

```python
df = df[['Survived', 'Pclass', 'Sex', 'Age', 'Fare']]
```

```python
df
```

```python
df['Age']= df['Age'].fillna(df['Age'].median())
```

```python
df['Sex']= df['Sex'].map({'male':1, 'female':0})
print(df.head())
print(df.isnull().sum())
```

```python
X = df[['Pclass', 'Sex', 'Age', 'Fare']]
Y = df['Survived']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
print("Training size:",  X_train.shape)
print("Testing size:", X_test.shape)
```

```python
#fluid and train the model
model = DecisionTreeClassifier(random_state=42)
model.fit(X_train, Y_train)
```

```python
#test model
predictions = model.predict(X_test)
#check accuracy
accuracy = accuracy_score(Y_test, predictions)
print("Model Accuracy ", round(accuracy * 100,2), "%")
```

```python
# Let's predict - will this person survive?
# Pclass=3, Sex=male(1), Age=22, Fare=7.25
person = [[3, 1, 22, 7.25]]
result = model.predict(person)

if result[0] == 1:
    print("This person SURVIVES! ✅")
else:
    print("This person does NOT survive ❌")
```

```python
# Pclass=3, Sex=male(1), Age=22, Fare=7.25
person = [[3, 0, 22, 7.25]]
result = model.predict(person)

if result[0] == 1:
    print("This person SURVIVES! ✅")
else:
    print("This person does NOT survive ❌")
```

```python
# Person 1 - Rich female, 1st class
person1 = [[1, 0, 25, 100]]
print("Person 1:", model.predict(person1))

# Person 2 - Poor male, 3rd class
person2 = [[3, 1, 25, 7]]
print("Person 2:", model.predict(person2))
```

```python
# Person 1 - Male, 3rd class, age 22
person1 = [[3, 1, 22, 7.25]]

# Person 2 - Female, 3rd class, age 22
person2 = [[3, 0, 22, 7.25]]

print("Male 3rd class:", model.predict(person1))
print("Female 3rd class:", model.predict(person2))

# Now try 1st class female
person3 = [[1, 0, 22, 100]]
print("Female 1st class:", model.predict(person3))
```

```python
from sklearn.ensemble import RandomForestClassifier

# Random Forest - stronger than Decision Tree
#Decision Tree = one person making a decision
#Random Forest = 100 people voting on a decision — majority wins!
#n_estimators= 100 means -- 100 trees coting..
model2 = RandomForestClassifier(n_estimators=100, random_state=42)
model2.fit(X_train, Y_train)

predictions2 = model2.predict(X_test)
accuracy2 = accuracy_score(Y_test, predictions2)

print("Decision Tree accuracy:", round(accuracy * 100, 2), "%")
print("Random Forest accuracy:", round(accuracy2 * 100, 2), "%")
```
